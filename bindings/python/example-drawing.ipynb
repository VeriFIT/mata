```python
import mata
import matplotlib.pyplot as plt
import networkx as nx
import scipy
import graphviz
import IPython
mata.configuration.store()['alphabet'] = mata.alphabets.EnumAlphabet(["0", "1"])
```

```python
lhs = mata.divisible_by(4)
lhs.label = "lhs"
rhs = mata.divisible_by(3)
rhs.label = "rhs"
result = mata.Nfa.concatenate(lhs, rhs)
result.label = "lhs.rhs"
```

```python
e_h = [
    (lambda aut, edge: edge.symb == 1, {'color': 'red'}),
    (lambda aut, edge: mata.configuration.Condition.transition_is_cycle, {'style': 'dashed', 'color': 'grey'})
]
n_h = [
    (lambda aut, state: state % 2 == 0, mata.configuration.Style.colored('red')),
    (lambda aut, state: state % 2 == 1, mata.configuration.Style.colored('blue')),
]
mata.plot(lhs, rhs, result, with_scc=True, node_highlight=n_h, edge_highlight=e_h)
```

```python
e_h = []
n_h = [
    (lambda aut, state: any(len(aut.post_of([state], i)) > 1 for i in (0, 1)), {'fillcolor': 'red', 'color': 'black', 'style': 'filled'}),
]
mata.plot(result, node_highlight=n_h, edge_highlight=e_h)
```
